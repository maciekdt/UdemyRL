```python
from sympy import *
from sympy.plotting import plot3d
from sympy.plotting.plot import Plot
```

```python
init_printing(use_latex='mathjax')
```

### Sigmoid function  
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

```python
x = symbols('x')
sigmoid = 1/(1 + exp(-x))
df = diff(sigmoid)
plot(sigmoid)
display(df)
plot(df)
```

### Hyperbolic tangent 
$$\tanh(x) = \frac{\sinh(x)}{\cosh(x)} = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$

```python
x = symbols('x')
tanh = (exp(x) - exp(-x))/(exp(x) + exp(-x))
df = diff(tanh)
plot(tanh)
display(df)
plot(df)
```

### ReLU function
$$\text{ReLU}(x) = \max(0, x)$$

```python
x = symbols('x')
relu = Max(0, x)
plot(relu)
df = diff(relu)
display(df)
plot(df)
```

### Leaky ReLU function
$$\begin{cases} 
x & \text{if } x > 0 \\
\alpha x & \text{if } x \leq 0 
\end{cases}$$

```python
x = symbols('x')
alpha = symbols('alpha', real=True, positive=True)
leaky_relu = Piecewise((x, x > 0), (alpha * x, x <= 0))

plot(leaky_relu.subs(alpha, 0.1))
```

```python
df = diff(leaky_relu.subs(alpha, 0.1))
display(df)
plot(df)
```
