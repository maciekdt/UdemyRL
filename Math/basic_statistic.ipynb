```python
from scipy.stats import binom, beta, norm
import matplotlib.pyplot as plt
import numpy as np
import random
```

### Rozkład dwumianowy (Bernouliego)
Czyli prawdopodobna liczba sukcesów k w n próbach przy prawdopodobieństwie sukcesu p

Wizualizacja dla 10 prób przy prawdopodobieństwie sukcesu 0.9.

```python
n = 10
p = 0.9

for k in range(n+1):
    probability = binom.pmf(k, n, p)
    print("{0} - {1}".format(k, probability))
```

```python
x = range(n+1)
probabilities = binom.pmf(x, n, p)

plt.bar(x, probabilities)
plt.show()
```

### Rozkład beta
Prawdopodobieństwo różnych bazowych prawdopodobieństw sukcesu, przy alfa sukcesach i beta niepowodzeniach. Funkcja CDF czyli kumulacyjna funkcja gęstości zwana dystrybuantą, zwaraca pole obszaru do danej wartości x.

W tym przypadku to jest prawdopodobieństwo, że dla zadanej liczby sukcesów i niepowodzeń prawdopodobieństwo bazowe jest mniejsze niż 0.9

```python
a = 8
b = 2

beta.cdf(.90, a, b)
```

Aby przedstawić na wykresie funkcję rozkładu beta należy skorzytsac z funkcji PDF gęstości rozkładu, która zwraca rzeczywistą wartość rozkładu. Pole pod wykresem tej funkcji zawsze jest równe 1.

```python
x = np.linspace(0, 1, 100)
pdf_values = beta.pdf(x, a, b)

plt.plot(x, pdf_values)
plt.show()
```

### Rozkład normalny
Zależy od średniej oraz odchylenia standardowego

```python
mean = 27.76
deviation = 5.23

x = np.linspace(7, 50, 100)
norm_values = norm.pdf(x, mean, deviation)

plt.plot(x, norm_values)
plt.show()
```

Dystrybuanta dla rozkładu normalnego. Prawdopodobieństwo na wartość mniejszą niż x.

```python
norm.cdf(23, mean, deviation)
```

Dystrybuanta odwrotna PPF, dla danego prawdopodobieństwa zwraca wartość. Dla poniższego przykładu ustala, że 90% próbek ma wartość mniejsząniż 34.46...

```python
norm.ppf(0.9, mean, deviation)
```

Generator liczb z rozkładu normalnego za pomocą odwrotnej dystrybuanty PPF.

```python
mean = 20.
deviation = 2.

for i in range(10):
    rand = random.uniform(0.0, 1.0)
    value = norm.ppf(rand, mean, deviation)
    print(value)
```
