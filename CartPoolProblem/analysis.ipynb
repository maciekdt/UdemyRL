```python
from trainer import Trainer
import matplotlib.pyplot as plt
```

```python
trainer = Trainer(alpha_rate=.2, gamma=.9, rbf_centers_size=100, epsilon=.2)
rewards_history = trainer.train(episodes_number=30_000, interval=1000, cache_file_path="q_cache.pkl")

plt.figure(figsize=(15, 10))
plt.plot(rewards_history)

plt.title("Avarage cumulative rewards in each episodes")
plt.xlabel("Episode")
plt.ylabel("Avarage reward")

plt.show()
```
