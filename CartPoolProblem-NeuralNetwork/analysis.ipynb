```python
from trainer import Trainer
from policy import Policy
from q_value_model import QValueModel
import matplotlib.pyplot as plt
```

```python
q_value_model = QValueModel(alpha_rate=.01, gamma=.95)
policy = Policy(train_mode=True, init_epsilon=1., min_epsilon=.01, delta_epsilon=.00005, q_value=q_value_model)
trainer = Trainer(policy=policy, q_value_model=q_value_model)
rewards_history = trainer.train(episodes_number=30_000, interval=1000, cache_file_path="model.pkl")

plt.figure(figsize=(15, 10))
plt.plot(rewards_history)

plt.title("Avarage cumulative rewards in each episodes")
plt.xlabel("Episode")
plt.ylabel("Avarage reward")

plt.show()
```
